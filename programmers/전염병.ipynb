# 전염병

**문제 설명**

m x n 크기인 사무실이 있습니다. 사무실에는 전염병에 걸린 직원이 있는데, 이 직원은 매일 상하좌우로 병을 퍼트려 다른 직원을 감염시킵니다. 단, 백신을 접종한 직원은 면역력이 있어 감염되지 않습니다.

예를 들어 2x4 크기 사무실에서, 병에 걸린 직원의 위치가 (1,4), (2,2)이고 백신을 맞은 직원의 위치가 (1,2)입니다. 이때 백신을 맞은 직원을 제외한 모든 직원이 병에 감염되기 까지는 이틀이 소요됩니다.

### 0일


### 1일


### 2일



사무실의 크기 `m`, `n`과 병에 걸린 직원의 위치 `infests`, 백신을 맞은 직원의 위치 `vaccinateds`가 매개변수로 주어집니다. 이때 백신을 맞은 직원을 제외한 모든 직원이 병에 감염되기까지 며칠이 걸리는지 return 하는 solution 함수를 완성해주세요.

**제한사항**

- m과 n은 1 이상 300 이하인 자연수입니다.
- infests의 길이는 1 이상 m * n 이하입니다.
    - infests의 원소는 [a, b] 형식이며, 1 ≤ a ≤ m, 1 ≤ b ≤ n입니다.
    - infests에는 같은 원소가 두 번 이상 들어있지 않습니다.
- vaccinateds는 길이가 1 이상 m * n 이하입니다.
    - vaccinateds의 원소는 [a, b] 형식이며, 1 ≤ a ≤ m, 1 ≤ b ≤ n입니다.
    - vaccinateds에는 같은 원소가 두 번 이상 들어있지 않습니다.
- 백신을 맞은 직원이 병에 걸린 경우는 주어지지 않습니다.
- 병을 아무리 퍼트려도 백신을 맞은 직원을 제외한 모든 직원이 병에 감염될 수 없다면 -1을 리턴합니다.

**입출력 예**

|m|	n|	infests	|vaccinateds|	result|
|:-:|:-:|:-|:-|:-:|
|2|	4|	[[1,4],[2,2]]	|[[1,2]]|	2|
|2|	3|	[[2,2]]	|[[1,2],[2,1],[2,3]]	|-1|
|2|	2|	[[1, 1], [2, 2]]	|[[1, 2], [2, 1]]|	0|

**입출력 예 설명**

- 입출력 예#1

    문제의 예시와 같습니다.


- 입출력 예#2


    전염병자의 근처가 백신 접종자로 막혀있어, 다른 직원에게 병이 도달하지 않습니다. 따라서 -1을 리턴합니다.
    
    
    
- 입출력 예#3


    처음부터 모든 직원이 전염병을 앓고 있거나, 백신을 맞은 상태입니다. 따라서 0을 리턴합니다.

- 2차원 리스트에서 m과 n 값의 인덱스가 1 크기 때문에 이것을 주의해야함

- 전염 리스트에 day를 어떻게 넣을것인지


    - 백신 위치를 어떻게 지도에 초기화할 것인지
        - 초기화 하는 대신 다음 방문 튜플이 vacinated에 있으면 이동 x


- python에서 2차원 이상의 리스트를 초기화 할땐
    - map_ = [[float('inf')] * n for _ in range(m)]
    - 만약 map_ = [[float('int') * n] * m 으로 초기화 할 경우 [float('inf') * n]은 같은 객체로 인식되기 때문이다



- 지도의 다음 전염위치를 어떻게 표시할 것인지


- day를 어떻게 증가시킬 것인지, 현재 day와 다음날 day의 차이를 두어야함

```python
from collections import deque

def solution(m, n, infests, vaccinateds):

    directions = [(0,-1),(0,1),(-1,0),(1,0)]
    day = 0
    map_ = [[float('inf')] * (n+1) for _ in range(m+1)] # 2 x 4 크기 지도
    count = 0

    # 백신 위치 초기화
    for x,y in vaccinateds:
        map_[x][y] = -1
    
    queue = deque(map(tuple, infests))

    while queue:
    
        for _ in range(len(queue)):
            x, y = queue.popleft()

            # 다음 전염 가능한 위치를 큐에 삽입
            if 0 < x <= m and 0 < y <= n and map_[x][y] != -1:
                count += 1
                map_[x][y] = -1

                queue.extend([(x + dx, y + dy) for dx, dy in directions])
            
        if queue:
            day += 1
            
    if count != m * n - len(vaccinateds):
        return -1
        
    return day - 1
```

```python
# Test Case 1
m = 2
n = 4
infests = [[1,4], [2,2]]
vaccinateds = [[1,2]]
assert 2 == solution(m, n, infests, vaccinateds)
```

```python
# Test Case 2
m = 2
n = 3
infests = [[2,2]]
vaccinateds = [[1,2],[2,1],[2,3]]
assert -1 == solution(m, n, infests, vaccinateds)
```

```python
# Test Case 3
m = 2
n = 2
infests = [[1, 1], [2, 2]]
vaccinateds = [[1, 2], [2, 1]]
assert 0 == solution(m, n, infests, vaccinateds)
```
