```python
from itertools import permutations

def solution(number, k):
    min_val = 0
    
    cand_list = list(permutations(number, k))
    
    case = set()
    for cand in cand_list:
        temp = list(number)
        for num in cand:
            temp.remove(num)
        case.add(int("".join(temp)))
    return str(max(case))
```
