# 가장큰수

**문제설명**

0 또는 양의 정수가 주어졌을 때, 정수를 이어 붙여 만들 수 있는 가장 큰 수를 알아내 주세요.

예를 들어, 주어진 정수가 [6, 10, 2]라면 [6102, 6210, 1062, 1026, 2610, 2106]를 만들 수 있고, 이중 가장 큰 수는 6210입니다.

0 또는 양의 정수가 담긴 배열 numbers가 매개변수로 주어질 때, 순서를 재배치하여 만들 수 있는 가장 큰 수를 문자열로 바꾸어 return 하도록 solution 함수를 작성해주세요.

**제한사항**
- numbers의 길이는 1 이상 100,000 이하입니다.
- numbers의 원소는 0 이상 1,000 이하입니다.
- 정답이 너무 클 수 있으니 문자열로 바꾸어 return 합니다.

**입출력 예**

|numbers|return|
|:---:|:---:|
|[6, 10, 2]|	"6210"|
|[3, 30, 34, 5, 9]|	"9534330"|

```python
# 오답1
from itertools import combinations

def solution(numbers):
    
    cand_list = [ cand for cand in list(permutations(numbers, len(numbers)))]
    cand_list = ["".join(str(cand).lstrip("(").rstrip(")").split(',')).replace(" ", "") for cand in cand_list]
    
    return str(max(cand_list))
```

```python
number_case_1 = [6,10,2]
solution(number_case_1)
```

```python
number_case_2 = [3,30,34,5,9]
solution(number_case_2)
```

- 나머지 연산으로 문제해결 시도 (divmod)
- 맨 앞의 숫자가 가장 큰 문자를 찾기 위해 문자의 아스키코드 값으로 정렬
- 앞의 숫자가 같지만 나머지가 0으로 떨어지는 경우 길이가 가장 작은 숫자를 기준으로 오름차순 정렬(3, 30, 300)
- 커스텀 정렬 https://pearlluck.tistory.com/599

```python
# 오답 2
from collections import deque

def solution(numbers):

    numbers = [str(num) for num in numbers]
    answer = []
    while True:
        if not numbers:
            break
        
        numbers = sorted(numbers, reverse=True)
        num = numbers.pop(0)
        answer.append(num)
    return "".join(answer)
```

```python
# 오답3 
def solution(numbers):
    numbers.sort(key=lambda x: str(x) * 3, reverse=True)
    return ''.join(list(map(str, numbers)))
```

```python
# 답안1
# 전체가 0인경우 
def solution(numbers):
    if len(numbers) == numbers.count(0):
        return '0'
    numbers.sort(key=lambda x: str(x) * 3, reverse=True)
    return ''.join(list(map(str, numbers)))
```

```python
number_case_1 = [6,10,2]
solution(number_case_1)
```

```python
number_case_2 = [3,30,34,5,9]
solution(number_case_2)
```

```python
# 답안 2
def solution(numbers):
    # 단순 정렬로는 안 됨, 30, 31, 32는 3보다 후순위임.
    # 998 1000 10 9 98 이라면 9(999)/998/98(98)/10/1000
    # 1000을 제외하면 최대 3자리까지 비교하면 되므로 * 3
    # Edge Case로 모든 원소가 0인 경우 0000000000000000000은 0이된다.
    if len(numbers) == numbers.count(0):
        return '0'
    return ''.join(sorted(list(map(str, numbers)), key=lambda x: x * 3, reverse=True))
```
