```python
from itertools import permutations

def solution(number, k):
    min_val = 0
    
    cand_list = list(permutations(number, k))
    
    case = set()
    for cand in cand_list:
        temp = list(number)
        for num in cand:
            temp.remove(num)
        case.add(int("".join(temp)))
    return str(max(case))
```

```python
number = "1924"
k = 2
assert 94 == solution(number, k)
```

```python
number = "1231234"
k = 3
assert 3234 == solution(number, k)
```

```python
number = "4177252841"
k = 4
assert 775841 == solution(number, k)
```
