```python
def dfs(start):

    visited[start] = True

    for adj in graph[start]:
        if not visited[adj]:
            dfs(adj)

n = int(input()) # 
m = int(input()) # 

graph = [[] for i in range(n+1)]
visited = [0 for i in range(n+1)]

for i in range(m):
    v1, v2 = map(int, input().split())
    graph[v1].append(v2)
    graph[v2].append(v1)

dfs(1)

print(sum(visited[2:]))
```
