```python
# dfs        

def dfs(x,y) :
    
    global cnt
    #     
    #  x,y  0 -1     N,M  x,y  
    if x <= -1 or x >= N or y <= -1 or y >= N :
        return False

    #       
    if graph[x][y] == 1:
        cnt += 1
        graph[x][y] = 0 #  0   

        # ,,,    
        dfs(x - 1, y)
        dfs(x, y - 1)
        dfs(x + 1, y)
        dfs(x, y + 1)
        return True
    return False

N = int(input()) # N * N 

# 2     
graph = []
g_lst = []
for i in range(N) :
    graph.append(list(map(int, input()))) #   0,1     

cnt = 0
for i in range(N) :
    for j in range(N):

        #   dfs 
        if dfs(i,j) == True:
            g_lst.append(cnt)
            cnt = 0

print(len(g_lst))
g_lst.sort()
for i in g_lst:
    print(i)
```
